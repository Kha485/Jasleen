## A) GET DATA FROM CSV FILE USING reader() function.

```python
import csv

# Load data from CSV
def load_data(file_path):
    data = []
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)
        for row in csv_reader:
            data.append(row)
    return header, data

# Load data
header, data = load_data('students_data.csv')
```

### a1. Retrieve the sex, age, number of relative in a family, state, and race based on the ID.

```python
def retrieve_by_id(data):
    print("COM728\nCOM728\nCOM728")
    student_id = input("Please enter the student ID: ").strip().upper()
    found = False
    for row in data:
        if row[0] == student_id:
            print(f"Sex: {row[1]}, Age: {row[2]}, Family Relatives: {row[21]}, State: {row[26]}, Race: {row[27]}")
            found = True
            break
    if not found:
        print(f"No student found with ID {student_id}")

retrieve_by_id(data)
```

### a2. Retrieve the sex, school support, access internet, attendance rate, and parental involvement associated with a specific race.

```python
def retrieve_by_race(data):
    print("COM728\nCOM728\nCOM728")
    race = input("Please enter the specific race: ")
    found = False
    for row in data:
        if row[27] == race:
            print(f"Sex: {row[1]}, School Support: {row[14]}, Access Internet: {row[19]}, Attendance Rate: {row[31]}, Parental Involvement: {row[37]}")
            found = True
        if not found:
            print(f"No student found with ID {race}")

retrieve_by_race(data)
```

### a3. Retrieve the ID, free time, math score, reading score and writing score of students whose absences are less than 50 based on the parental involvement.

```python
def retrieve_by_absences(data, max_absences):
    print("COM728\nCOM728\nCOM728")
    for row in data:
        if int(row[25]) < max_absences:
            print(f"ID: {row[0]}, Freetime: {row[22]}, Math Score: {row[28]}, Reading Score: {row[29]}, Writing Score: {row[30]}")

retrieve_by_absences(data, 50)
```

### a4. Retrieve information from your chosen columns and apply a specific condition that relates to an individual student.

```python
def retrieve_by_condition(data):
    print("COM728\nCOM728\nCOM728")
    for row in data:
        # Checking if student is ina relation and its Math Score is greater than 85 and internet is accessed or not
        if int(row[28]) > 85 and str(row[20]).strip().lower() == "yes":
            print(f"ID: {row[0]}, Internet Access: {row[19]}, Math Score: {row[28]}, Reading Score: {row[29]}, Writing Score: {row[30]}")
            continue

retrieve_by_condition(data)
```

### b1. Identify the top 3 levels of mothers education for a specific race of students based on the parental involvement levels.

```python
def top_mother_education(df, race, parental_involvement):
    print("COM728\nCOM728\nCOM728")
    filtered_df = df[(df['Race'] == race) & (df['Parental_involvement'] == parental_involvement)]
    top_education = filtered_df['Mother_education_level '].value_counts().head(3)
    print(top_education)

top_mother_education(df, 'Asian', 'low')
```

### b2. Analyse the average number of absences among students with a particular level of parental involvement.

```python
def student_avg_absences(df, parental_involvement):
    filtered_df = df[df['Parental_involvement'] == parental_involvement]
    avg_absences = filtered_df['Absences'].mean()
    print("COM728\nCOM728\nCOM728")
    print(f"Average Absences for {parental_involvement} parental involvement: {avg_absences}")

student_avg_absences(df, 'low')
```

## C) Visualise data using the matplotlib module

```python
import matplotlib.pyplot as plt
```

### c1. Create a chart to illustrate the proportion of students in this school based on their race.

```python
race_counts = df['Race'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(race_counts.index, race_counts.values, color=['skyblue','cyan','red','pink'])
plt.title('Proportion of Students by Race')
plt.xlabel('Race')
plt.ylabel('Number of Students')
plt.show()
```

### c2. Create a chart to visually compare the average writing scores among students in each race group.

```python
avg_scores = df.groupby('Race')['Writing_score'].mean()
plt.figure(figsize=(10, 5))
plt.bar(avg_scores.index, avg_scores.values, color='green')
plt.title('Average Writing Scores by Race')
plt.xlabel('Race')
plt.ylabel('Average Writing Score')
plt.show()
```

### c3. Create a chart to illustrate the relationship between students' reading and writing scores.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Reading_score'], df['Writing_score'], color=['violet'])
plt.title('Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(df['Reading_score'], df['Writing_score'], gridsize = 30, cmap ='Purples', mincnt=1)
plt.colorbar(label = 'Count')
plt.title('Relationship between students Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

### c4. Create a visualisation of your selection to showcase information related to student performance that can reveal trends, different, behaviours, or patterns, ensuring it is distinct from previous requirements.

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
activities_vs_scores = df.groupby('Activities')[['Math_score', 'Reading_score', 'Writing_score']].mean()
studytime_vs_scores = df.groupby('Studytime')[['Math_score', 'Reading_score', 'Writing_score']].mean()
health_vs_scores = df.groupby('Health')[['Math_score', 'Reading_score', 'Writing_score']].mean()

# Activities vs Scores
activities_vs_scores.plot(kind='bar', ax=axs[1], color=['blue', 'orange', 'green'])
axs[1].set_title('Activities vs Scores')
axs[1].set_xlabel('Activities (No: 0, Yes: 1)')
axs[1].set_ylabel('Average Score')

# Study Time vs Scores
axs[0].plot(studytime_vs_scores.index, studytime_vs_scores['Math_score'], label='Math Score', marker='o')
axs[0].plot(studytime_vs_scores.index, studytime_vs_scores['Reading_score'], label='Reading Score', marker='o')
axs[0].plot(studytime_vs_scores.index, studytime_vs_scores['Writing_score'], label='Writing Score', marker='o')
axs[0].set_title('Study Time vs Scores')
axs[0].set_xlabel('Study Time')
axs[0].set_ylabel('Average Score')
axs[0].legend()

# Health vs Scores
axs[2].plot(health_vs_scores.index, health_vs_scores['Math_score'], label='Math Score', marker='o')
axs[2].plot(health_vs_scores.index, health_vs_scores['Reading_score'], label='Reading Score', marker='o')
axs[2].plot(health_vs_scores.index, health_vs_scores['Writing_score'], label='Writing Score', marker='o')
axs[2].set_title('Health vs Scores')
axs[2].set_xlabel('Number of Failures')
axs[2].set_ylabel('Average Score')
axs[2].legend()

plt.tight_layout()
plt.show()
```
